```python
import numpy as np

x = np.array([1, 2, 3, 4, 5, 1, 2, 1, 1, 1])

# Get unique values and their counts
values, counts = np.unique(x, return_counts=True)

# Find the index of the maximum count
max_count_index = np.argmax(counts)

# Most frequent value
most_frequent_value = values[max_count_index]

print("Most frequent value:", most_frequent_value)

# Find indices where this value occurs
indices = np.where(x == most_frequent_value)[0]
print("Indices of most frequent value:", indices)
```

```python
import numpy as np 

n_array=np.array([[55,25,15],[30,44,2],[11,45,77]])


# 1. Sum of diagonal elements (trace)
diag_sum = np.trace(n_array)
print("Sum of diagonal elements:", diag_sum)

# 2. Eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(n_array)
print("\nEigenvalues:\n", eigenvalues)
print("\nEigenvectors (each column corresponds to an eigenvector):\n", eigenvectors)

# 3. Inverse of the matrix
try:
    inv_matrix = np.linalg.inv(n_array)
    print("\nInverse of the matrix:\n", inv_matrix)
except np.linalg.LinAlgError:
    print("\nMatrix is singular and cannot be inverted.")

# 4. Determinant of the matrix
determinant = np.linalg.det(n_array)
print("\nDeterminant of the matrix:", determinant)
```

```python
import numpy as np

p=np.array([[1,2],[2,3]])
q=np.array([[4,5],[6,7]])


# Matrix multiplication
product = np.dot(p, q)  # or p @ q

print("Product of p and q:\n", product)

#To find the covariance matrix, you need a dataset where rows represent variables and columns 
#represent observations (or vice versa). Here, you can stack the arrays p and q as vectors and compute covariance:

# Flatten the matrices to 1D arrays (vectors)
p_flat = p.flatten()
q_flat = q.flatten()

# Stack the vectors as rows
data = np.vstack((p_flat, q_flat))

# Compute covariance matrix
cov_matrix = np.cov(data)

print("\nCovariance matrix:\n", cov_matrix)

```

```python
import numpy as np

x=np.array([2,3,4],[3,2,9])
y=np.array([1,5,0],[5,10,3])

# 1. Inner Product (Dot Product)
inner_product = np.dot(x, y.T)  # y.T is the transpose of y
print("Inner Product (Dot Product):\n", inner_product)

# 2. Outer Product
# For outer product, we treat both vectors as 1D arrays
outer_product = np.outer(x.flatten(), y.flatten())  # Flattening to 1D vectors
print("\nOuter Product:\n", outer_product)

# 3. Cartesian Product
# Cartesian product involves combining each element of x with each element of y.
# np.meshgrid is used for Cartesian product.
cartesian_product = np.array(np.meshgrid(x.flatten(), y.flatten())).T.reshape(-1, 2)
print("\nCartesian Product:\n", cartesian_product)
```

```python
import numpy as np

# Correct way to define 2D arrays
x = np.array([[2, 3, 4], [3, 2, 9]])  # Two rows and three columns
y = np.array([[1, 5, 0], [5, 10, 3]])  # Two rows and three columns

# 1. Inner Product (Dot Product)
inner_product = np.dot(x, y.T)  # y.T is the transpose of y
print("Inner Product (Dot Product):\n", inner_product)

# 2. Outer Product
# For outer product, we treat both vectors as 1D arrays
outer_product = np.outer(x.flatten(), y.flatten())  # Flattening to 1D vectors
print("\nOuter Product:\n", outer_product)

# 3. Cartesian Product
# Cartesian product involves combining each element of x with each element of y.
# np.meshgrid is used for Cartesian product.
cartesian_product = np.array(np.meshgrid(x.flatten(), y.flatten())).T.reshape(-1, 2)
print("\nCartesian Product:\n", cartesian_product)
```

```python
import numpy as np 

array = np.array([[1, -2, 3],[-4, 5, -6]]) 

# i) element-wise absolute value
abs_array = np.abs(array)
print("absolute value:\n",abs_array)

# ii)
# Flattened array
flattened = array.flatten()
percentiles_flat = np.percentile(flattened, [25, 50, 75])
print('Flattened array:\n',percentiles_flat)

# Percentiles for each column
percentiles_cols = np.percentile(array, [25, 50, 75], axis=0)
print('Percentiles for each column:\n',percentiles_cols)

# Percentiles for each row
percentiles_rows = np.percentile(array, [25, 50, 75], axis=1)
print('Percentiles for each row:\n',percentiles_rows)

# iii)
# Flattened
mean_flat = np.mean(flattened)
print('mean of flattened:\n',mean_flat)

median_flat = np.median(flattened)
print('median of flattened:\n',median_flat)

std_flat = np.std(flattened)
print('standard deviation of flattened:\n',std_flat)

# Per column
mean_col = np.mean(array, axis=0)
print('mean per column:\n',mean_col)

median_col = np.median(array, axis=0)
print('median per column:\n',median_col)

std_col = np.std(array, axis=0)
print('standard deviation per column:\n',std_col)

# Per row
mean_row = np.mean(array, axis=1)
print('mean per row:\n',mean_row)

median_row = np.median(array, axis=1)
print('median per row:\n',median_row)

std_row = np.std(array, axis=1)
print('standard deviation per row:\n',std_row)
```

```python
import numpy as np

a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0])

floor_values=np.floor(a)
print("Floor values",floor_values)

ceil_values=np.ceil(a)
print("Ceiling values",ceil_values)

truncated_values=np.trunc(a)
print('Truncated values',truncated_values)

rounded_values=np.round(a)
print('Rounded values',rounded_values)
```

```python
import numpy as np

arr = np.array([10, 52, 62, 16, 16, 54, 453])

sorted_array=np.sort(arr)
print('sorted array:\n',sorted_array)

indices_of_sorted=np.argsort(arr)
print('indices of sorted array:\n',indices_of_sorted)

smallest_4 = np.sort(array)[:4]
print('smallest 4:\n',smallest_4)

largest_5 = np.sort(array)[-5:]  # ascending order
print("5 Largest elements:\n", largest_5)
```
